```python
import pandas as pd
```

```python
housing_data = pd.read_csv('Resource/GrossVacancyRates75LargestMetro2005to2019.csv')
housing_data = housing_data.rename(columns={'Unnamed: 0':'Metro Area','Unnamed: 1': '2005','Unnamed: 2': '2006', 'Unnamed: 3': '2007',
                                           'Unnamed: 4': '2008', 'Unnamed: 5': '2009', 'Unnamed: 6': '2010',
                                           'Unnamed: 7': '2011', 'Unnamed: 8': '2012', 'Unnamed: 9': '2013',
                                           'Unnamed: 10': '2014', 'Unnamed: 11': '2015', 'Unnamed: 12': '2016',
                                           'Unnamed: 13': '2017', 'Unnamed: 14': '2018', 'Unnamed: 15': '2019'})
housing_data = housing_data.iloc[2: , :]
housing_data.drop(housing_data.tail(2).index,inplace=True)
display (housing_data)
```

```python
housing_change_pct = (((housing_data['2005'])-housing_data['2019']) / ((housing_data['2005'] + housing_data['2019']) /2) *100) 
housing_data['Percent of Change']=housing_change_pct
housing_data = housing_data.sort_values(by='Percent of Change', ascending=True)
housing_data['Percent of Change']=housing_data['Percent of Change'].map('{:.2f}%'.format)
pd.set_option("display.max_rows", None, "display.max_columns", None)
display(housing_data)
```

Some of the data from this set matches the more limited set of rental vacancies. Metro areas like Albany-Schenectady-Troy have retained their place. The huge differences apparent in housing as an aggregate vs. housing as specifically rental may imply that different pressures are fueling the change in vacancies between different housing markets. If not, I would expect the top 10 to be relatively homogeneous between the two.
