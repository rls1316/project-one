```python
#LB
#Dependencies 
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
#METRO_VACANCY
#load metro vacancy rates 2005 - 2014
metro_vacancy05_14 = pd.read_csv("../Resource/metrovacancy05to19.csv")
metro_vacancy05_14.head()
```

```python
#METRO_VACANCY
#load metro vacancy rates 2015 - 2019
metro_vacancy15_19 = pd.read_csv("../Resource/metrovacancy15_19.csv")
metro_vacancy15_19.head()
```

```python
#METRO_VACANCY
metro_vacancy05_19 = metro_vacancy05_14.merge(metro_vacancy15_19, how='inner')
metro_vacancy05_19.head()
```

```python
#METRO_VACANCY
Pct_change = (((metro_vacancy05_19['2019']) - metro_vacancy05_19['2005'])/metro_vacancy05_19['2005'])  * 100
metro_vacancy05_19["Pct_change"] = Pct_change
#sort
metro_vacancy05_19.sort_values(by = ["Pct_change"], inplace = True, ascending = False)
#format%
metro_vacancy05_19["Pct_change"] = metro_vacancy05_19["Pct_change"].map('{:.2f}%'.format)
metro_vacancy05_19.head()
```

```python
#METRO_VACANCY
Top_10_metro_vacancy05_19 = metro_vacancy05_19.head(10)
Top_10_metro_vacancy05_19
```

```python
#METRO_VACANCY
Top_10_metro_vacancy05_19.plot(x="NAME", y=["2005", "2010", "2015", "2019"], kind='bar',figsize=(9,8))
plt.xticks(rotation=90)
plt.title("Top 10 Metro Area by Gross Vacancy Rate within the Top 75 Metro Areas, 5 year basis")
plt.xlabel("Top Metro Area")
plt.ylabel("Homeowner Vacancy Rate %")
plt.tight_layout()
```

```python
#METRO_VACANCY
Top_10_metro_vacancy05_19.plot(x = "NAME", y=["2005"], kind = 'scatter')
plt.xticks(rotation=90)
Top_10_metro_vacancy05_19.plot(x = "NAME", y=["2019"], kind = 'scatter')
plt.xticks(rotation=90)
```

scatter isnt really working to tell any type of correlation at this point
