```python
import pandas as pd

```

```python
#load vacancy data 
vacancies_data = pd.read_csv("Resource/RentalVacancyRates75LargestMetro2015to2019.csv")
vacancies_data = vacancies_data.iloc[1: , :]
#for some reason the last row is all nan, dropped it
vacancies_data.drop(vacancies_data.tail(1).index,inplace=True)
#dropped columns are nan values left over because the csv is a boilerplate with too many year columns
vacancies_data = vacancies_data.drop(['Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], axis = 1)
display(vacancies_data)
```

```python
#rename columns to years
vacancies_data = vacancies_data.rename(columns={'Unnamed: 1': '2015', 'Unnamed: 2': '2016', 'Unnamed: 3': '2017', 'Unnamed: 4': '2018', 'Unnamed: 5': '2019'})
#determine percent of change from 2015-19, store in column
vacancy_change_pct = (((vacancies_data['2015'])-vacancies_data['2019']) / ((vacancies_data['2015'] + vacancies_data['2019']) /2) *100) 
vacancies_data['Percent Change']=vacancy_change_pct
#vacancies_data['Percent Change']=vacancies_data['Percent Change'].abs()
vacancies_data = vacancies_data.sort_values(by='Percent Change', ascending=True)
vacancies_data['Percent Change']=vacancies_data['Percent Change'].map('{:.2f}%'.format)
pd.set_option("display.max_rows", None, "display.max_columns", None)
display(vacancies_data)
```

In this df, negative percentage indicates an increase in vacancy, while positive indicates less vacancies. The metro areas involved with the greatest increase in vacancies are surprisingly not found in the counties that experienced the most population growth.
