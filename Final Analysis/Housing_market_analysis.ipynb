```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read in housing document
housingdf = pd.read_csv('Resource/housing_w_dc.csv')
```

```python
housingdf = housingdf.drop(columns='Unnamed: 0', axis = 1)
```

```python
housingdf
```

```python
#grouped by state, county and year to show data 
housing_group = housingdf.groupby(['State','County','Year']).sum()
```

```python
housing_group.head(50)
```

```python
#picked county to look at data
idaho = housingdf.loc[(housingdf['County']=='Greenlee County')]
```

```python
idaho
```

```python
#plotted data of housing market
idaho.plot.line('Year',['Total Vacancies','Total Renter Occupied','Owner Occupied','For Rent','For Sale'],figsize=(9,8))
plt.title('Greenlee County, AZ housing market')
plt.ylabel('Housing numbers')
```

```python
idaho.plot.line('Year',['Total Housing'],figsize=(9,8))
plt.title('Greenlee County, AZ housing supply')
```
