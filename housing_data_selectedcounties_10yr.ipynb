```python
year = []
state=[]
totalhousing = []
totalvacancy =[]
totalrenteroccupied = []
owneroccupied = []
forrent = []
forsale = []
county = []
```

```python
ds="https://api.census.gov/data/2012/acs/acs5"
```

```python
for url in base_urls : 
    for x in params : 
        house = requests.get(url,params=x).json()
        
        try : 
            county.append(house[1][0].split(",")[0])
            totalhousing.append(house[1][1])
            totalvacancy.append(house[1][2])
            totalrenteroccupied.append(house[1][3])
            owneroccupied.append(house[1][4])
            forrent.append(house[1][5])
            forsale.append(house[1][6])
            state.append(house[1][0].split(",")[1])
            year.append(url.split("/")[4])
        except : 
            print('woops')
       # print(x)
```

```python
housing_df['State']= state
```

```python
housing_df['Total Housing'] = totalhousing
housing_df['Total Vacancies'] = totalvacancy
housing_df['Total Renter Occupied'] = totalrenteroccupied
housing_df['Owner Occupied'] = owneroccupied
housing_df['For Rent']= forrent
housing_df['For Sale']=forsale
             
```

```python
housing_df['County'] = county
```
