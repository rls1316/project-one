```python
#Dependencies 
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Load Population Data set into pandas
Population_Data = pd.read_csv("../Resource/US_Population_Estimates_2010_2019.csv")
Population_Data.head()
```

```python
#Identify Incomplete Rows
Population_Data.count()
```

```python
#Calculate the % difference by County and add to DF with format
Pct_Growth = (((Population_Data['2019']) - Population_Data['Estimate Base (4/1/2010)'])/Population_Data['Estimate Base (4/1/2010)']) * 100
Population_Data["Pct Growth"] = Pct_Growth

#Sort DF by highest Pct of Growth
Population_Data.sort_values(by = ["Pct Growth"], inplace = True, ascending = False)

#Add formatting to the Pct Growth
Population_Data["Pct Growth"] = Population_Data["Pct Growth"].map('{:.2f}%'.format)

Population_Data['State'] = Population_Data['State']

Population_Data.head()
```

```python
#Consolidate the data to State level totals for each year
State_Population_Data = Population_Data.groupby(['State']).sum()
State_Population_Data.reset_index()
State_Population_Data.head()
```

```python
#Calculate the % difference and add to DF with format
Pct_Growth = (((State_Population_Data['2019']) - State_Population_Data['Estimate Base (4/1/2010)'])/State_Population_Data['Estimate Base (4/1/2010)']) * 100
State_Population_Data["Pct Growth"] = Pct_Growth

#Sort DF by highest Pct of Growth
State_Population_Data.sort_values(by = ["Pct Growth"], inplace = True, ascending = False)

#Add formatting to the Pct Growth
State_Population_Data["Pct Growth"] = State_Population_Data["Pct Growth"].map('{:.2f}%'.format)

State_Population_Data = State_Population_Data.rename_axis(index = '', columns = 'State')

State_Population_Data.head(10)
```

```python
#New DataFrame of the Top 10 States for Pct Growth of Population
Top_10_State_Population = State_Population_Data.head(10)
Top_10_State_Population.reset_index(inplace = True)

Top_10_State_Population
```

```python
#Get the top 5 counties for each State
Top_Counties_by_State = Population_Data.loc[Population_Data.State.isin(Top_10_State_Population[''])]

Top_Counties_by_State
```
