```python
import requests
import pandas as pd 
from api import census 
from pprint import pprint
```

```python
base_urls=["https://api.census.gov/data/2010/acs/acs5","https://api.census.gov/data/2011/acs/acs5",
"https://api.census.gov/data/2012/acs/acs5","https://api.census.gov/data/2013/acs/acs5",
"https://api.census.gov/data/2014/acs/acs5","https://api.census.gov/data/2015/acs/acs5",
"https://api.census.gov/data/2016/acs/acs5","https://api.census.gov/data/2017/acs/acs5",
"https://api.census.gov/data/2018/acs/acs5","https://api.census.gov/data/2019/acs/acs5"]
```

```python
year = []
state=[]
totalhousing = []
totalvacancy =[]
totalrenteroccupied = []
owneroccupied = []
forrent = []
forsale = []
county = []
```

```python
dc={'for':'state:11','get' : 'NAME,B25001_001E,B25004_001E,B25003_003E,B25003_002E,B25004_002E,B25004_004E','key' : census}
```

```python
dc1 = requests.get(base_urls[1], params=dc).json()

dc1
```

```python
for url in base_urls :
    dc_house=requests.get(url,params=dc).json()
    county.append(dc_house[1][0].split(",")[0])
    totalhousing.append(dc_house[1][1])
    totalvacancy.append(dc_house[1][2])
    totalrenteroccupied.append(dc_house[1][3])
    owneroccupied.append(dc_house[1][4])
    forrent.append(dc_house[1][5])
    forsale.append(dc_house[1][6])
    state.append(dc_house[1][0].split(",")[0])
    year.append(url.split("/")[4])
```

```python
dc_df = pd.DataFrame({'Year': year})
```

```python
dc_df['State']= state

dc_df['Total Housing'] = totalhousing
dc_df['Total Vacancies'] = totalvacancy
dc_df['Total Renter Occupied'] = totalrenteroccupied
dc_df['Owner Occupied'] = owneroccupied
dc_df['For Rent']= forrent
dc_df['For Sale']=forsale
```
