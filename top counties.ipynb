```python
counties = pd.read_csv('Resource/top100counties.csv')
```

```python
counties_chosen=counties[counties['Geographic Area'].str.contains('Texas|Utah|Colorado|Florida|Nevada|Idaho|Arizona|North Dakota|Washington|District of Columbia')]
```

```python
counties_chosen=counties_chosen.rename(columns={'Unnamed: 3':'population estimate 2019','Unnamed: 5' : 'Percent Change'})
counties_chosen
```

```python
states = ['Texas','Utah','Colorado','Florida','Nevada','Idaho','Arizona','North Dakota','Washington']

for state in states:
    top_counties = counties_chosen[counties_chosen['Geographic Area'].str.contains(state)]
    if state == 'Texas':
        toptex = pd.DataFrame(top_counties)
    elif state == 'Utah':
        toputah = pd.DataFrame(top_counties)
    elif state == 'Colorado':
        topcol = pd.DataFrame(top_counties)
    elif state == 'Florida':
        topflor = pd.DataFrame(top_counties)
    elif state == 'Nevada':
        topnev = pd.DataFrame(top_counties)
    elif state == 'Idaho':
        topida = pd.DataFrame(top_counties)
    elif state == 'Arizona':
        toparz = pd.DataFrame(top_counties)  
    elif state == 'North Dakota':
        topndk = pd.DataFrame(top_counties)
    elif state == 'Washington':
        topwa = pd.DataFrame(top_counties)
```
